### Define a function to pull the country data from the API

```python
def get_country_data(country):
    """
    Function to get data about a country from "https://api.covid19api.com" API
    """
    country_name=str(country)
    url = service_url + country_name 
    print(url)
    try:
        uh = urllib.request.urlopen(url)
    except HTTPError as e:
        print("Sorry! Could not retrive anything on {}".format(country_name))
        return None
    except URLError as e:
        print('Failed to reach a server.')
        print('Reason: ', e.reason)
        return None
    else:
        data = uh.read().decode()
        print("Retrieved data on {}. Total {} characters read.".format(country_name,len(data)))
        return data
    print(url)
```

```python
country_name = 'South-Korea'
```

```python
import urllib.parse
query = urllib.parse.quote(country_name)
```

```python
data=get_country_data(query)
```
